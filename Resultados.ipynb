Importações

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Resultados PC 1

```python
files1 = ['./results/results-PC1-V1-1000rows.csv', './results/results-PC1-V2-1000rows.csv', './results/results-PC1-V3-1000rows.csv', './results/results-PC1-V4-1000rows.csv', './results/results-PC1-V5-1000rows.csv', 
          './results/results-PC1-V6-1000rows.csv', './results/results-PC1-V7-1000rows.csv', './results/results-PC1-V8-1000rows.csv', './results/results-PC1-V9-1000rows.csv', './results/results-PC1-V10-1000rows.csv']

dataframes  = []
for file in files1:
    df = pd.read_csv(file)
    dataframes.append(df)

results1= pd.concat(dataframes)


mean_execution_times = results1.groupby('Threads')['Execution Time'].mean().reset_index()
threads1 = mean_execution_times['Threads']
execution_times1 = mean_execution_times['Execution Time']

# Tempo de execução com um único thread
T1 = execution_times1[threads1 == 1].values[0]

# Calcular Speed-Up e Eficiência
speed_up1 = T1 / execution_times1
efficiency1 = speed_up1 / threads1

# Plotar o gráfico de Speed-Up
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(threads1, speed_up1, marker='o', linestyle='-', color='b')
plt.title('Speed-Up em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Speed-Up')
plt.grid(True)

# Plotar o gráfico de Eficiência
plt.subplot(1, 2, 2)
plt.plot(threads1, efficiency1, marker='o', linestyle='-', color='r')
plt.title('Eficiência em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Eficiência')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Resultados PC 2

```python
files2 = ['./results/results-PC2-V1-1000rows.csv', './results/results-PC2-V2-1000rows.csv', './results/results-PC2-V3-1000rows.csv', './results/results-PC2-V4-1000rows.csv', './results/results-PC2-V5-1000rows.csv', 
          './results/results-PC2-V6-1000rows.csv', './results/results-PC2-V7-1000rows.csv', './results/results-PC2-V8-1000rows.csv', './results/results-PC2-V9-1000rows.csv', './results/results-PC2-V10-1000rows.csv']
dataframes  = []
for file in files2:
    df = pd.read_csv(file)
    dataframes.append(df)

results2 = pd.concat(dataframes)

mean_execution_times = results2.groupby('Threads')['Execution Time'].mean().reset_index()
threads2 = mean_execution_times['Threads']
execution_times2 = mean_execution_times['Execution Time']

# Tempo de execução com um único thread
T1 = execution_times2[threads2 == 1].values[0]

# Calcular Speed-Up e Eficiência
speed_up2 = T1 / execution_times2
efficiency2 = speed_up2 / threads2

# Plotar o gráfico de Speed-Up
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(threads2, speed_up2, marker='o', linestyle='-', color='b')
plt.title('Speed-Up em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Speed-Up')
plt.grid(True)

# Plotar o gráfico de Eficiência
plt.subplot(1, 2, 2)
plt.plot(threads2, efficiency2, marker='o', linestyle='-', color='r')
plt.title('Eficiência em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Eficiência')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Compação

```python
# Plotar o gráfico comparativo de Speed-Up
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(threads1, speed_up1, marker='o', linestyle='-', color='b', label='PC1')
plt.plot(threads2, speed_up2, marker='o', linestyle='-', color='g', label='PC2')
plt.title('Speed-Up em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Speed-Up')
plt.legend()
plt.grid(True)

# Plotar o gráfico comparativo de Eficiência
plt.subplot(1, 2, 2)
plt.plot(threads1, efficiency1, marker='o', linestyle='-', color='b', label='PC1')
plt.plot(threads2, efficiency2, marker='o', linestyle='-', color='g', label='PC2')
plt.title('Eficiência em Função do Número de Threads')
plt.xlabel('Número de Threads')
plt.ylabel('Eficiência')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
